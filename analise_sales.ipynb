# Análise de Dados do Dataset Vendas Supermercado

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando o dataset sales.csv
sales = pd.read_csv('sales.csv')
sales.head()
```

```python
# Lista nome colunas
print(sales.columns)
```

```python
# Informações gerais do dataset
sales.info()
```

```python
# Análise exploratória dos dados
# Distribuição das vendas por produto
plt.figure(figsize=(10,6))
sns.countplot(x='product_name', data=sales)
plt.title('Distribuição das Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Contagem de Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Análise exploratória dos dados
# Distribuição das vendas por cidade
plt.figure(figsize=(10,6))
sns.countplot(x='city', data=sales, color="#134e79")
plt.title('Distribuição das Vendas por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Contagem de Vendas')
plt.xticks(rotation=45)
plt.show()
```
